Problem #15: In-place vs Regular Operations
Create an array and demonstrate the difference between in-place and regular broadcasting operations.

# Expected Output:
Regular operation: new array created
In-place operation: original array modified
Memory addresses differ for regular, same for in-place

```python
import numpy as np

a = np.array([1, 2, 3], dtype=int)
b = np.array([10, 20, 30], dtype=int)

print("Before any op")
print("a =", a)
print("a id =", id(a))
print("a data ptr = ", a.ctypes.data)
print()

c = a + b   
print("Regular operation: c = a + b")
print("a after regular op:", a)           
print("c = ", c)
print("a id = ", id(a))
print("c id = ", id(c))                      
print("a data ptr = ", a.ctypes.data)
print("c data ptr = ", c.ctypes.data)      
print()

d = np.array([1, 2, 3], dtype=int)         
ptr_before = d.ctypes.data
print("Before in-place: d id = ", id(d), "data ptr:", ptr_before)

d += b    
ptr_after = d.ctypes.data
print("In-place operation =  d += b")
print("d after in-place = ", d)
print("d id after in-place = ", id(d))       
print("data ptr equal = ", ptr_before == ptr_after)  
print("d data ptr = ", ptr_after)
```

# Expected Output:
[[15 18]
 [15 18]]

```python
import numpy as np 

arr = np.array([[1, 2],
                [3, 4]])
arr2 = np.array([[5,6]])

adv_op = np.where(arr > arr2, arr * 2, arr2 * 3)

print(adv_op)
```

Problem #20: Comprehensive Workflow
Create arrays, perform: reshape â broadcast â arithmetic â indexing. Track all attribute changes.
