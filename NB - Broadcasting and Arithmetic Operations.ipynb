Problem #17: Broadcasting with Transpose
Create a 3x1 array and a 1x3 array. Multiply them after transposing one to make broadcasting work correctly.

# Expected Output:
[[ 0  0  0]
 [ 0  1  2]
 [ 0  2  4]]

```python
import numpy as np 

arr1 = np.arange(3*1).reshape(3,1)
arr2 = np.arange(1*3).reshape(1,3)

#TRANSPOSE arr1 
trans_arr1 = np.transpose(arr1)
trans_arr2 = np.transpose(arr2)
multi_op = trans_arr1 * trans_arr2

print(multi_op)
```

Problem #20: Comprehensive Broadcasting Example
Create a 2x3 array and a 3-element 1D array. Perform addition, multiplication, and division operations between them using broadcasting and print all results.

# Expected Output:
Addition:
[[ 1  3  5]
 [ 4  6  8]]
Multiplication:
[[ 0  2  6]
 [ 0  5 12]]
Division:
[[0.  2.  1.5]
 [0.  5.  3. ]]

```python
import numpy as np 

arr = np.arange(1, 2 * 3 + 1).reshape(2,3)
arr2 = np.arange(3)

add_op = arr + arr2 
multi_op = arr * arr2 

mask = arr2 > 0

div_op = np.divide(arr, arr2, where = mask)


print("Addtion:")
print(add_op)
print("Multiplication:")
print(multi_op)
print("Division:")
print(div_op)
```
