Problem #20: Comprehensive Array Operations
Create an array [5, 2, 5, 8, 2, 9]. Filter to keep values greater than 4, find unique values, sort them, and then concatenate with [1, 2, 3].

# Expected Output:
Filtered (>4): [5, 5, 8, 9]
Unique: [5, 8, 9]
Sorted: [5, 8, 9]
Final concatenated: [5, 8, 9, 1, 2, 3]


```python
import numpy as np

arr = np.array([5, 2, 5, 8, 2, 9])
conct_arr = np.array([1, 2, 3])

filter_4 = arr[np.where(arr > 4)]
unique_values = np.unique(filter_4)
sorting = np.sort(unique_values)
final_concate = np.concatenate((sorting, conct_arr))

print(f"""Filtered (>4): {filter_4}
Unique: {unique_values}
Sorted: {sorting}
Final concatenated: {final_concate}""")
```

Problem #20: Comprehensive DateTime Workflow
Create datetime array for a week â†’ filter weekdays â†’ attach random sales data â†’ sort by sales â†’ display top 3.


# Expected Output:
Top 3 sales days:
[['2023-01-04' 95]
 ['2023-01-03' 87]
 ['2023-01-02' 76]]

```python
import numpy as np 

all_dates = np.arange('2023-01-01', '2023-01-08' ,dtype = 'datetime64[D]')

filter_weekdays = np.is_busday(all_dates)

week_days = all_dates[filter_weekdays]

# for sales:

ran_sales = np.random.default_rng(42)
sales = ran_sales.integers(50, 101, size = week_days.shape[0])

sorting = np.argsort(sales)[::-1]
top_3indx = sorting[:3]

 
top_3_sales = sales[top_3indx]
top_3_dates = week_days[top_3indx]


table = np.column_stack((top_3_dates.astype('datetime64[D]').astype(str), top_3_sales.astype(str)))

print("\nTop 3 sales days:")
print(table) # the output is not exactly the same as there is random function used
```
