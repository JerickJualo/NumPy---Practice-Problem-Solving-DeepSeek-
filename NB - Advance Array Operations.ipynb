Problem #1: Basic Axis Swapping
Create a 2x3 array with values 1-6. Swap the axes (transpose) and print both original and swapped arrays.

# Expected Output:
Original:
[[1 2 3]
 [4 5 6]]
Swapped:
[[1 4]
 [2 5]
 [3 6]]

```python
import numpy as np 

arr = np.arange(1, 7).reshape(2,3)

trans_arr = np.transpose(arr)

print(f"Original: \n {arr}")
print(f"Swapped: \n {trans_arr}")
```

Problem #3: Vertical Array Joining
Create two 2x3 arrays: [[1, 2, 3], [4, 5, 6]] and [[7, 8, 9], [10, 11, 12]]. Join them vertically.

Problem #14: Axis Swapping with 3D Array
Create a 2x2x3 array with values 1-12. Swap axes 0 and 1, then print the new shape.

# Expected Output:
Original shape: (2, 2, 3)
After swap: (2, 2, 3)  # Same shape but different memory layout

```python
import numpy as np 

arr = np.arange(1, 13).reshape(2,2,3)

swaps = np.swapaxes(arr, 0, 1)

print(f"Original shape and strides: {arr.shape}, {arr.strides}")
print(f"After swap shape and strides: {swaps.shape}, {swaps.strides} # Same shape but different memory layout")
```

Problem #20: Comprehensive Array Operations
Create an array [5, 2, 5, 8, 2, 9]. Filter to keep values greater than 4, find unique values, sort them, and then concatenate with [1, 2, 3].

# Expected Output:
Filtered (>4): [5, 5, 8, 9]
Unique: [5, 8, 9]
Sorted: [5, 8, 9]
Final concatenated: [5, 8, 9, 1, 2, 3]


```python
import numpy as np

arr = np.array([5, 2, 5, 8, 2, 9])
conct_arr = np.array([1, 2, 3])

filter_4 = arr[np.where(arr > 4)]
unique_values = np.unique(filter_4)
sorting = np.sort(unique_values)
final_concate = np.concatenate((sorting, conct_arr))

print(f"""Filtered (>4): {filter_4}
Unique: {unique_values}
Sorted: {sorting}
Final concatenated: {final_concate}""")
```
