```python
import numpy as np 
import time

c_array = np.array(np.arange(1000000).reshape(1000,1000), order = 'c')

f_array = np.array(np.arange(1000000).reshape(1000,1000), order = 'f')


start_time = time.time()
for _ in range(100):  
    c_sum = np.sum(c_array, axis=1)
end_time = time.time()
c_time = (end_time - start_time) / 100


start_time = time.time()
for _ in range(100):
    f_sum = np.sum(f_array, axis=1)
end_time = time.time()
f_time = (end_time - start_time) / 100

print(f"C-ordered time: {c_time:.4f}s")
print(f"F-ordered time: {f_time:.4f}s")
```

Problem #19: Stride Impact on Operations
Compare row-wise vs column-wise operations on C-order and F-order arrays and observe relative performance.

python
# Expected Output:
C-order: row operations faster
F-order: column operations faster


```python
import numpy as np
import timeit

rows, cols = 1000, 1000

c_order_array = np.array(np.random.rand(rows, cols), order='C')
f_order_array = np.array(np.random.rand(rows, cols), order='F')

c_order_row_time = timeit.timeit("c_order_array.sum(axis=1)", globals=globals(), number=100)
c_order_col_time = timeit.timeit("c_order_array.sum(axis=0)", globals=globals(), number=100)

f_order_row_time = timeit.timeit("f_order_array.sum(axis=1)", globals=globals(), number=100)
f_order_col_time = timeit.timeit("f_order_array.sum(axis=0)", globals=globals(), number=100)

print(f"C-order Array:")
print(f"Row-wise sum time: {c_order_row_time:.6f} seconds")
print(f"Column-wise sum time: {c_order_col_time:.6f} seconds")
print(f"Result: {'row operations faster' if c_order_row_time < c_order_col_time else 'column operations faster'}")

print(f"\nF-order Array:")
print(f"Row-wise sum time: {f_order_row_time:.6f} seconds")
print(f"Column-wise sum time: {f_order_col_time:.6f} seconds")
print(f"Result: {'row operations faster' if f_order_row_time < f_order_col_time else 'column operations faster'}")
```
