Problem #14: Fancy Indexing Shape
Create a 5x5 array. Use fancy indexing with indices [0, 2, 4] for both rows and columns and print the shape of the result.

# Expected Output:
(3, 3)

```python
import numpy as np 

arr = np.arange(5*5).reshape(5,5)
row_indices = np.array([0, 2, 4])
column_indices = np.array([0, 2, 4])
fancy = arr[np.ix_(row_indices, column_indices)]

print(fancy.shape)
```

Problem #18: Array Manipulation and Shape Preservation
Create a 3x3 array. Flatten it and then check if the original array shape is preserved.

# Expected Output:

Problem #19: Memory Layout Strides
Create a 3x4 array in C-order (row-major) and F-order (column-major). Print the strides for each.

# Expected Output:
C-order strides: (32, 8)
F-order strides: (8, 24)

```python
import numpy as np 

arr_C = np.full((3,4), 4, order = 'C')
arr_F = np.full((3,4), 5, order = 'F')

print(f"C-order strides: {arr_C.strides}")
print(f"F-order strides: {arr_F.strides}")
```
