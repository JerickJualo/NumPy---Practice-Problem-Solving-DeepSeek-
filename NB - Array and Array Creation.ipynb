Problem #6: Multi-axis Concatenation

Problem #7: Flattening Order Matters
Create a 2x3 array [[1, 2, 3], [4, 5, 6]].

# Expected Output (C-order):

[1 2 3 4 5 6]
# Expected Output (F-order):

[1 4 2 5 3 6]

```python
import numpy as np

array_1 = np.array([[1, 2, 3], [4, 5, 6]])

c_flat = array_1.flatten(order = 'C')
f_flat = array_1.flatten(order = 'F')

print(c_flat)
print(f_flat)
```
