Problem #6: Forcing a Type

Create a 1D array from the list [1, 2, 3, 4] but explicitly set the data type to float32. Print the array and its dtype.

Expected Output:

[1. 2. 3. 4.]
float32

Solution Below:

```python
import numpy as np

array = np.array([1,2,3,4], dtype = np.float32)

print(array)
print(array.dtype)
```

Problem #8: The Transpose
Create a 2x4 array from a range of 8 numbers, reshape it, and then print its transpose.

Expected Output (Original 2x4):

[[0 1 2 3]
 [4 5 6 7]]

Expected Output (Transpose):

[[0 4]
 [1 5]
 [2 6]
 [3 7]]

Solution Below:

```python
import numpy as np 

array = np.arange(8).reshape(2,4)

transposed_array =  np.transpose(array)

print(array, "\n")
print(transposed_array)
```

Problem #9: Flatten it Out

Create a 2x3 array and then flatten it (convert it to a 1D array). Print both the original and the flattened array.

Expected Output (Original):

[[10 20 30]
 [40 50 60]]

Expected Output (Flattened):

[10 20 30 40 50 60]

Solution Below:

```python
import numpy as np 

array = np.arange(10,61,10).reshape(2,3)
print(array,"\n")

flatten_array = array.flatten()
print(flatten_array)
```

Problem #11: Stacking Vertically

Create two 1D arrays: a = [1, 2, 3] and b = [4, 5, 6]. Stack them vertically to form a 2x3 array and print it.

Expected Output:

[[1 2 3]
 [4 5 6]]

Solution Below:

```python
import numpy as np

array_1 = np.array([1 ,2, 3])
array_2 = np.array([4, 5, 6])

array_3 = np.concatenate((array_1, array_2)).reshape(2,3)
print(array_3)
```
